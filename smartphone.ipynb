```python
#IMPORTING THE PANDAS 

import pandas as pd
```

```python
#IMPORTING THE DATASET CSV FILE.

df = pd.read_csv(r"smartphones.csv")
```

```python
#GETTING A CONCISE SUMMARY OF THE DATAFRAME. 
#USING .info()

df.info()
```

```python
#GETTING THE STATISTICAL SUMMARY OF THE DATAFRAME.
#USING .describe()

df.describe()
```

```python
#TO RETURN THE LAST 'n' ROWS OF A DATAFRAME (DEFAULT 5 ROWS)


df.tail()
```

```python
#CHECKING THE NUMBER OF ROWS AND COLOUMNS


df.shape
```

```python
#CHECKING WHERE VALUES ARE MISSING (TRUE/FALSE)


df.isnull()
```

```python
#COUNTING MISSING VALUES PER COLOUMN


df.isnull().sum()
```

```python
#TOTAL MISSING VALUES IN THE DATASET


df.isnull().sum().sum()
```

```python
#CHECK DATA TYPES


df.dtypes
```

```python
#FILLING MISSING RATINGS WITH MEAN


df['rating'] = df['rating'].fillna(df['rating'].mean())


```

```python
#FILLING MISSING CATEGORICAL COLOUMNS WITH MODE


for col in ['camera', 'card', 'os']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#VERIFYING


print("\nMissing values after handling:\n", df.isnull().sum())
```

```python
#CHECKING NUMBER OF DUPLICATE ROWS


print("Duplicate rows before:", df.duplicated().sum())
```

```python
#REMOVING DUPLICATES


df.drop_duplicates(inplace=True)
```

```python
#CHECKING AGAIN


print("Duplicate rows after:", df.duplicated().sum())
```

```python
#CLEAN AND CONVERT NUMERIC COLOUMNS STORED AS TEXT
```

```python
#REMOVING ₹ AND COMMAS , CONVERTING PRICE TO FLOAT


df['price'] = df['price'].replace('[₹,]', '', regex=True).astype(float)
```

```python
#EXTRACTING NUMERIC RAM IN GB


df['ram'] = df['ram'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
#EXTRACTING BATTERY CAPACITY IN MAH


df['battery'] = df['battery'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
#EXTRACTING DISPLAY SIZE IN INCHES


df['display'] = df['display'].astype(str).str.extract(r'(\d+\.\d+)').astype(float)
```

```python
#EXTRACTING MAIN CAMERA MEGAPIXELS


df['camera'] = df['camera'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
#VERIFYING DATATYPES


df.dtypes
```

```python
#ENSURING THE DATASET IS CLEAN AND READY FOR ANALYSIS
```

```python
#STANDARDISE COLOUMN NAMES


df.columns = df.columns.str.strip().str.lower()
```

```python
#FINAL VALIDATION


print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
print("\nPreview cleaned data:\n", df.head())
```

```python
#DATA SET IS READY FOR ANALYSIS

#FOR SAVING THE CLEANED DATA SET   ===>
                         #OPTIONAL


df.to_csv("smartphones_cleaned.csv", index=False)
```
